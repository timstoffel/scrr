```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import push_notebook, show, output_notebook, export_png
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.models.glyphs import VArea
from bokeh.models.tools import HoverTool
import bokeh.palettes
from bokeh.transform import factor_cmap

import pandas as pd

df = pd.read_csv('Results/ab-export-2020-01-09-16-31-53.csv')
df = df.replace(-1, 0)
print(df)
```

```python
# Block f√ºr node-express
filter_node = df[df['image_name'] == 'node-express']
print(filter_node)
```

```python
grouped_node = filter_node.groupby(['runtime', 'number_of_targets', 'repeat'])['requests_per_second'].sum().reset_index(name='sum_requests_per_second')
print(grouped_node)

grouped_sum = grouped_node.groupby(['runtime', 'number_of_targets'])['sum_requests_per_second'].mean().reset_index(name='mean_requests_per_second')
print(grouped_sum)

grouped_min = grouped_node.groupby(['runtime', 'number_of_targets'])['sum_requests_per_second'].min().reset_index(name='min_requests_per_second')
print(grouped_min)

grouped_max = grouped_node.groupby(['runtime', 'number_of_targets'])['sum_requests_per_second'].max().reset_index(name='max_requests_per_second')
print(grouped_max)
```

```python
# Filter auf default

default_sum = grouped_sum[grouped_sum['runtime'] == 'default']
default_sum = default_sum[['number_of_targets', 'mean_requests_per_second']]
print(default_sum)

default_min = grouped_min[grouped_min['runtime'] == 'default']
default_min = default_min[['number_of_targets', 'min_requests_per_second']]
print(default_min)

default_max = grouped_max[grouped_max['runtime'] == 'default']
default_max = default_max[['number_of_targets', 'max_requests_per_second']]
print(default_max)

default_data = pd.merge(default_sum, default_min, on='number_of_targets')
default_data = pd.merge(default_data, default_max, how='outer', on='number_of_targets')
print(default_data)


# Filter auf kata

kata_sum = grouped_sum[grouped_sum['runtime'] == 'kata']
kata_sum = kata_sum[['number_of_targets', 'mean_requests_per_second']]
print(kata_sum)

kata_min = grouped_min[grouped_min['runtime'] == 'kata']
kata_min = kata_min[['number_of_targets', 'min_requests_per_second']]
print(kata_min)

kata_max = grouped_max[grouped_max['runtime'] == 'kata']
kata_max = kata_max[['number_of_targets', 'max_requests_per_second']]
print(kata_max)

kata_data = pd.merge(kata_sum, kata_min, on='number_of_targets')
kata_data = pd.merge(kata_data, kata_max, how='outer', on='number_of_targets')
print(kata_data)

# Filter auf katafc

katafc_sum = grouped_sum[grouped_sum['runtime'] == 'katafc']
katafc_sum = katafc_sum[['number_of_targets', 'mean_requests_per_second']]
print(katafc_sum)

katafc_min = grouped_min[grouped_min['runtime'] == 'katafc']
katafc_min = katafc_min[['number_of_targets', 'min_requests_per_second']]
print(katafc_min)

katafc_max = grouped_max[grouped_max['runtime'] == 'katafc']
katafc_max = katafc_max[['number_of_targets', 'max_requests_per_second']]
print(katafc_max)

katafc_data = pd.merge(katafc_sum, katafc_min, on='number_of_targets')
katafc_data = pd.merge(katafc_data, katafc_max, how='outer', on='number_of_targets')
print(katafc_data)

# Filter auf runsc

runsc_sum = grouped_sum[grouped_sum['runtime'] == 'runsc']
runsc_sum = runsc_sum[['number_of_targets', 'mean_requests_per_second']]
print(runsc_sum)

runsc_min = grouped_min[grouped_min['runtime'] == 'runsc']
runsc_min = runsc_min[['number_of_targets', 'min_requests_per_second']]
print(runsc_min)

runsc_max = grouped_max[grouped_max['runtime'] == 'runsc']
runsc_max = runsc_max[['number_of_targets', 'max_requests_per_second']]
print(runsc_max)

runsc_data = pd.merge(runsc_sum, runsc_min, on='number_of_targets')
runsc_data = pd.merge(runsc_data, runsc_max, how='outer', on='number_of_targets')
print(runsc_data)
```

```python
default_source = ColumnDataSource(default_data)
print(default_source.data)

kata_source = ColumnDataSource(kata_data)
print(kata_source.data)

katafc_source = ColumnDataSource(katafc_data)
print(katafc_source.data)

runsc_source = ColumnDataSource(runsc_data)
print(runsc_source.data)
```
