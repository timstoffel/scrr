```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import push_notebook, show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.models.tools import HoverTool
import itertools 
import bokeh.palettes
from bokeh.transform import factor_cmap

import pandas as pd

df = pd.read_csv('Results/ab-export-2020-01-09-16-31-53.csv')
df = df.replace(-1, 0)
print(df)
```

```python
grouped = df.groupby(['image_name', 'runtime'])['requests_per_second'].mean()
print(grouped)

test = ColumnDataSource(pd.DataFrame(df))
print(test.data['runtime'])
```

```python
source = ColumnDataSource(pd.DataFrame(grouped))
print(source.data)
```

```python
runtimes = source.data['image_name_runtime'].tolist()

print(runtimes)

p = figure(x_range=FactorRange(*runtimes), width=1500)

colors = bokeh.palettes.d3['Category10'][5]
colors = colors + colors + colors
print(colors)

colors_itter = itertools.cycle(colors)

color_map = factor_cmap(field_name='image_name_runtime',
                    palette=colors, factors=runtimes)

p.vbar(x='image_name_runtime', top='requests_per_second', source=source, width=0.70, color=color_map)

p.title.text ='Schnitt Request pro Sekunde'
p.xaxis.axis_label = 'Runtime'
p.yaxis.axis_label = 'Request pro Sekunde'

output_file("request_per_secound_gouped_by_runtime_image.html")
output_notebook()

show(p)
```
